```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step_function(x):
    return np.array(x > 0, dtype=np.int)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    x = x - np.max(x, axis=-1, keepdims=True)
    return np.exp(x) / np.sum(np.exp(x), axis=-1, keepdims=True)
```

```python
x = np.arange(-5, 5, 0.01)
```

```python
def plot(x, activation, title):
    plt.plot(x, activation(x))
    plt.title(title)
    plt.grid()
    plt.savefig(title + ".png")
```

```python
plot(x, step_function, "step_function")
```

```python
plot(x, sigmoid, "sigmoid_function")
```

```python
plot(x, relu, "relu_function")
```

```python
plot(x, softmax, "softmax_function")
```
